## Criterion

**Criterion 1**

A function $ f $ is continuous at $ x_0 $ if and only if for every $ \epsilon > 0 $, there exists $ \delta > 0 $ such that $ |x - x_0| < \delta $ implies $ |f(x) - f(x_0)| < \epsilon $.
